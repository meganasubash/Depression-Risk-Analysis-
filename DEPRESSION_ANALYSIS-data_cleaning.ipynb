```python
import pandas as pd

df = pd.read_csv(r'C:\Users\MS994\Downloads\fds\depression_dataset_raw.csv')


df['How do you usually cope with stress?'] = (
    df['How do you usually cope with stress?']
    .astype(str)
    .str.replace('â€™', "'", regex=False)
    .str.replace('â€œ', '"', regex=False)
    .str.replace('â€�', '"', regex=False)
)





# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(r'[^\w]+', '_', regex=True)
)
```

```python
# Remove duplicate rows
df = df.drop_duplicates()

# Drop completely empty rows
df = df.dropna(how='all')

```

```python
print(df.isnull().sum())
```

```python
df = df.rename(columns={
    'how_often_do_you_feel_stressed_': 'stress',
    'how_often_do_you_feel_anxious_or_worried_': 'anxiety',
    'how_often_do_you_feel_sad_or_low_': 'sadness',
    'do_you_feel_overwhelmed_by_daily_tasks_': 'overwhelmed',
    'do_you_have_difficulty_concentrating_': 'concentration',
    'how_often_do_you_overthink_situations_': 'overthinking',
    'do_you_experience_mood_swings_': 'mood_swings',
    'how_many_hours_do_you_sleep_daily_': 'sleep_raw',
    'how_often_do_you_feel_tired_or_low_on_energy_': 'tired',
    'do_you_feel_supported_by_people_around_you_': 'support',
    'how_often_do_you_socialize_with_others_': 'social',
    'how_do_you_usually_cope_with_stress_': 'coping',
    'how_would_you_rate_your_overall_mental_well_being_': 'wellbeing'
})

df = df.apply(lambda col: col.astype(str).str.strip().str.capitalize())

```

```python
df['wellbeing'] = df['wellbeing'].astype(str).str.lower().str.strip()
wellbeing_map = {
    'very poor': 1,
    'poor': 2,
    'average': 3,
    'good': 4,
    'excellent': 5
}

df['wellbeing'] = df['wellbeing'].replace(wellbeing_map)

df['wellbeing'] = pd.to_numeric(df['wellbeing'], errors='coerce')

df['wellbeing'] = df['wellbeing'].fillna(df['wellbeing'].median())

df.head(10)
```

```python
import numpy as np
df.replace("Nan", np.nan, inplace=True)
```

```python
df['overwhelmed'] = df['overwhelmed'].replace({
    'Yes': 'Often',
    'No': 'Never'
})
```

```python
df['support'] = df['support'].replace({
    'Yes': 'Often',
    'No': 'Never'
})
```

```python
df['mood_swings'] = df['mood_swings'].replace({
    'Yes': 'Often',
    'No': 'Never'
})
```

```python
df['social'] = df['social'].replace({
    'Daily': 'Always',
    'Weekly': 'Often'
})
```

```python
def clean_sleep(val):
    val = str(val).lower()

    if 'less' in val:
        return 'Less than 5'
    elif 'more' in val:
        return 'More than 8'
    else:
        return '5-7'

df['sleep'] = df['sleep_raw'].apply(clean_sleep)
df=df.drop('sleep_raw',axis=1)
```

```python
df['coping'] = df['coping'].replace({
    'Music/entertainment': 'Healthy',
    'Talking to someone': 'Healthy',
    'Exercise': 'Healthy',
    'Sleeping': 'Neutral',
    'Avoiding problems': 'Avoidance',
    'Other': 'Neutral',
    'I don’t cope well': 'Avoidance'   # ← EXACT match
})
```

```python
df.fillna(df.mode().iloc[0], inplace=True)
```

```python
likert = {
    "Never": 0,
    "Rarely": 1,
    "Sometimes": 2,
    "Often": 3,
    "Always": 4
}

cols = ['stress','anxiety','sadness','overwhelmed',
        'concentration','overthinking','tired',
        'social','support','mood_swings']

for col in cols:
    df[col] = df[col].map(likert)
```

```python
df['sleep'] = df['sleep'].map({
    'Less than 5': 0,
    '5-7': 1,
    'More than 8': 2
})

df['coping'] = df['coping'].map({
    'Healthy': 1,
    'Neutral': 2,
    'Avoidance': 3
})
```

```python
df['emotional_score'] = df[['stress','anxiety','sadness']].mean(axis=1)

df['cognitive_score'] = df[['overthinking','concentration','overwhelmed']].mean(axis=1)

df['sleep_risk'] = 2 - df['sleep']
df['physical_score'] = (df['sleep_risk'] + df['tired']) / 2

df['social_score'] = (df['social'] + df['support']) / 2

df['behavioral_score'] = (df['mood_swings'] + df['coping']) / 2
```

```python
df[['emotional_score','cognitive_score','physical_score',
    'social_score','behavioral_score']].head()
```

```python
df['drs_raw'] = (
    df['emotional_score'] * 0.30 +
    df['cognitive_score'] * 0.25 +
    df['physical_score'] * 0.20 +
    df['behavioral_score'] * 0.10 -
    df['social_score'] * 0.15
)
```

```python
#normalize
df['drs'] = (df['drs_raw'] / 4) * 100
```

```python
def categorize(score):
    if score <= 33:
        return "Low"
    elif score <= 66:
        return "Moderate"
    else:
        return "High"

df['risk_level'] = df['drs'].apply(categorize)
```

```python
# Coping Score Meaning:
# 0 → Very poor coping (Avoidance, "I don't cope well")
#     - No effective coping mechanism
#     - High vulnerability to depression

# 1 → Low coping (Neutral / passive coping like sleeping, other)
#     - Temporary relief but not problem-solving

# 2 → Moderate coping (e.g., Music/Entertainment)
#     - Distraction-based coping, somewhat helpful

# 3 → Healthy coping (e.g., Talking to someone, Exercise)
#     - Active and effective coping strategies
#     - Reduces depression risk
```

```python
# Well-being Score Meaning:
# 1 → Very Poor (Severe distress, high depression risk)
# 2 → Poor (Low mental health)
# 3 → Average (Moderate mental state)
# 4 → Good (Stable mental health)
# 5 → Excellent (Very healthy mental state)

# Note:
# For depression scoring, this is usually inverted:
# depression_score = 6 - wellbeing
```

```python
# Frequency Scale Meaning:
# 0 → Never (No symptoms)
# 1 → Rarely (Very low occurrence)
# 2 → Sometimes (Moderate occurrence)
# 3 → Often (Frequent symptoms)
# 4 → Always (Severe / constant symptoms)
```

```python
# Sleep Score Meaning:
# 0 → Healthy sleep (7–8 hours)
# 1 → Slightly disturbed sleep
# 2 → Poor sleep
# 3 → Very poor sleep (<5 hours)
```

```python
# ================================
# DEPRESSION DATASET SCORING GUIDE
# ================================

# General Rule:
# Higher score → Higher depression severity
# Lower score → Better mental health

# -------------------------------
# CORE SYMPTOMS (0–4 scale)
# -------------------------------

# stress:
# 0 → Never stressed
# 1 → Rarely stressed
# 2 → Sometimes stressed
# 3 → Often stressed
# 4 → Always stressed

# anxiety:
# 0 → No anxiety
# 4 → Constant anxiety

# sadness:
# 0 → No sadness
# 4 → Persistent sadness

# overwhelmed:
# 0 → Never overwhelmed
# 4 → Always overwhelmed

# concentration:
# 0 → No difficulty
# 4 → Severe difficulty concentrating

# overthinking:
# 0 → No overthinking
# 4 → Constant overthinking

# mood_swings:
# 0 → Stable mood
# 4 → Frequent mood swings

# tired:
# 0 → Energetic
# 4 → Always tired / low energy

# -------------------------------
# SUPPORT & SOCIAL (0–4 scale)
# -------------------------------

# support:
# 0 → No support
# 4 → Strong support system
# NOTE: This should be inverted for scoring

# social:
# 0 → No social interaction
# 4 → Highly social
# NOTE: Invert for depression score

# -------------------------------
# COPING (0–3 scale)
# -------------------------------

# coping:
# 0 → Avoidance / "I don’t cope well"
# 1 → Passive coping (sleeping, other)
# 2 → Distraction coping (music, entertainment)
# 3 → Healthy coping (talking, exercise)
# NOTE: Invert for depression score

# -------------------------------
# WELL-BEING (1–5 scale)
# -------------------------------

# wellbeing:
# 1 → Very Poor mental health
# 2 → Poor
# 3 → Average
# 4 → Good
# 5 → Excellent
# NOTE: MUST be inverted → (6 - wellbeing)

# -------------------------------
# SLEEP (0–3 scale)
# -------------------------------

# sleep:
# 0 → Healthy sleep
# 1 → Slight disturbance
# 2 → Poor sleep
# 3 → Very poor sleep (<5 hrs)

# -------------------------------
# FINAL NOTE
# -------------------------------
# Positive factors (support, social, coping, wellbeing)
# should be inverted before calculating final depression score
```
